# Análisis

## Número de publicaciones por año

```python
# Prepare dataframe
my_plot = my_BOE.groupby('AÑO')['ID'].count()
# Make plot
fig, ax = plt.subplots(figsize=(16, 4.5))
my_plot.plot(ax=ax)
ax.set_xlabel('')
ax.set_ylabel('Número de publicaciones')
ax.set_title('Número de publicaciones en el BOE por año', fontsize=16, weight='bold')
plt.show()
```

## Número de páginas por año

```python
# Prepare dataframe
my_plot = my_BOE.groupby('AÑO')['PAGINAS'].sum()
# Make plot
fig, ax = plt.subplots(figsize=(16, 4.5))
my_plot.plot(ax=ax)
ax.set_xlabel('')
ax.set_ylabel('Número de páginas')
ax.set_title('Número de páginas publicadas en el BOE por año', fontsize=16, weight='bold')
plt.show()
```
